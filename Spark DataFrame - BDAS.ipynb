```python
df = pd.read_excel(r"Datasets/dataset_to_be_clean.xlsx",'dataset')
df.info()
```

```python
df = df.fillna(999)
df.info()
```

```python
df = df.drop("Nacionality", axis='columns')
df.info()
```

```python

df = df[(df["Curricular units 1st sem (enrolled)"] > 0)
   |
   (df["Curricular units 2nd sem (enrolled)"] > 0)]
uselesscolumns = [
    "Application mode",
    "Application order",
    "Tuition fees up to date",
    "Curricular units 1st sem (credited)",
    "Curricular units 1st sem (enrolled)",
    "Curricular units 1st sem (evaluations)",
    "Curricular units 1st sem (approved)",
    "Curricular units 2nd sem (credited)",
    "Curricular units 2nd sem (enrolled)",
    "Curricular units 2nd sem (evaluations)",
    "Curricular units 2nd sem (approved)",
    ]
for x in uselesscolumns:
    df = df.drop(x, axis='columns')
df["Total Grade"] = df["Curricular units 1st sem (grade)"] + df["Curricular units 2nd sem (grade)"]
df.info()
df
```

```python
df = df[df["Target"] != "Enrolled"]
g = df.groupby('Target')
bdf = g.apply(lambda x: x.sample(g.size().min()).reset_index(drop=True))
sns.countplot(bdf['Target'])
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("Pandas to Spark Linear Regression") \
    .getOrCreate()
spark_df_social_for_grade = spark.createDataFrame(bdf)
spark_df_personal_for_grade = spark.createDataFrame(bdf)

socialFeaturecols = [
    'Unemployment rate',
    'GDP'
    ]
personalFeaturecols = [
    'Daytime/evening attendance',
    'Educational special needs',
    'Scholarship holder',
    'Age at enrollment',
    'Previous qualification'
    ]
assembler_social_for_grade = VectorAssembler(inputCols=socialFeaturecols, outputCol='features')
spark_df_social_for_grade = assembler_social_for_grade.transform(spark_df_social_for_grade)
spark_df_social_for_grade_train, spark_df_social_for_grade_test = spark_df_social_for_grade.randomSplit([0.5, 0.5])

assembler_personal_for_grade = VectorAssembler(inputCols=personalFeaturecols, outputCol='features')
spark_df_personal_for_grade = assembler_personal_for_grade.transform(spark_df_personal_for_grade)
spark_df_personal_for_grade_train, spark_df_personal_for_grade_test = spark_df_personal_for_grade.randomSplit([0.5, 0.5])

print(spark_df_social_for_grade_train.show())
print(spark_df_personal_for_grade_train.show())
```

```python
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

lr_social = LinearRegression(featuresCol='features', labelCol='Total Grade').setRegParam(0.3)
lr_social_model = lr_social.fit(spark_df_social_for_grade_train)
lr_personal = LinearRegression(featuresCol='features', labelCol='Total Grade').setRegParam(0.3)
lr_personal_model = lr_personal.fit(spark_df_personal_for_grade_train)

predictions_social = lr_social_model.transform(spark_df_social_for_grade_test)
predictions_personal = lr_personal_model.transform(spark_df_personal_for_grade_test)

predictions_social.select('features', 'Total Grade', 'prediction').show()
predictions_personal.select('features', 'Total Grade', 'prediction').show()

evaluator = RegressionEvaluator(labelCol='Total Grade', predictionCol='prediction', metricName='rmse')

rmse_social = evaluator.evaluate(predictions_social)
print(rmse_social)
rmse_personal = evaluator.evaluate(predictions_personal)
print(rmse_personal)
```

```python
bdf['Target_numeric'], unique = pd.factorize(bdf['Target'])
spark_df_social_for_target = spark.createDataFrame(bdf)
spark_df_personal_for_target = spark.createDataFrame(bdf)
socialFeaturecols = [
    'Unemployment rate',
    'GDP',
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation"
    ]
personalFeaturecols = [
    'International',
    'Educational special needs',
    'Marital status',
    'Age at enrollment',
    'Previous qualification',
    'Debtor',
    'Gender',
    'Course'
    ]

assembler_social_for_target = VectorAssembler(inputCols=socialFeaturecols, outputCol='features')
spark_df_social_for_target = assembler_social_for_grade.transform(spark_df_social_for_target)
spark_df_social_for_target_train, spark_df_social_for_target_test = spark_df_social_for_target.randomSplit([0.5, 0.5])

assembler_personal_for_target = VectorAssembler(inputCols=personalFeaturecols, outputCol='features')
spark_df_personal_for_target = assembler_personal_for_grade.transform(spark_df_personal_for_target)
spark_df_personal_for_target_train, spark_df_personal_for_target_test = spark_df_personal_for_target.randomSplit([0.5, 0.5])

print(spark_df_social_for_target_train.show())
print(spark_df_personal_for_target_train.show())
```

```python
from pyspark.ml.classification import LinearSVC
from pyspark.ml.evaluation import BinaryClassificationEvaluator

lsvc_social = LinearSVC(featuresCol='features', labelCol='Target_numeric')
lsvc_social_model = lsvc_social.fit(spark_df_social_for_target_train)
lsvc_personal = LinearSVC(featuresCol='features', labelCol='Target_numeric')
lsvc_personal_model = lsvc_personal.fit(spark_df_personal_for_target_train)

predictions_social = lsvc_social_model.transform(spark_df_social_for_target_test)
predictions_personal = lsvc_personal_model.transform(spark_df_personal_for_target_test)

predictions_social.select('features', 'Target_numeric', 'prediction').show()
predictions_personal.select('features', 'Target_numeric', 'prediction').show()

evaluator = BinaryClassificationEvaluator(labelCol='Target_numeric', rawPredictionCol='prediction', metricName='areaUnderROC')

auc_social = evaluator.evaluate(predictions_social)
print(auc_social)
auc_personal = evaluator.evaluate(predictions_personal)
print(auc_personal)
```

```python
print(f"Coefficients: {lsvc_personal_model.coefficients}")
print(f"Intercept: {lsvc_personal_model.intercept}")
```
